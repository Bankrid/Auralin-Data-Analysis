#### Define 
* multiply the weight of the patient with patient_id 211 by 2.20462 to convert it from kilogram to pounds
* change the height of the patient patient_id 5 from 27 to 72

```python
# multiply the weight of the patient with patient_id 211 by 2.20462 to convert it from kilogram to pounds

patient_clean.loc[patient_clean.patient_id == 211, 'weight'] = round(patient_clean.loc[patient_clean.patient_id == 211, 'weight'] * 2.20462, 1)

# change the height of the patient patient_id 5 from 27 to 72
patient_clean.loc[patient_clean.patient_id == 5, 'height'] = 72
```

#### Define
* extract and save different phone number and email from contact and drop contact
* melt auralin and novodra columns into treatment, start_dose and end_dose 
* join adverse_clean with treatment clean
* remove given name and surname from treatment dataframe and then create patient_id as a common column

#### Code

```python
#extract email
patient_clean["email"] = patient_clean.contact.str.extract(r"([a-zA-Z][a-zA-Z0-9_+.-]+@[a-zA-Z0-9+._-]+\.[a-zA-Z0-9_+-.]+[a-zA-Z])")

#extract phone number 
patient_clean["phone_number"] = patient_clean["phone_number"] = patient_clean.contact.str.extract(r"((?:\+\d{1,2}\s)?\(?\d{3}\)?[\s.-]?\d+[\s.-]\d{4})")

# drop contact column
patient_clean.drop(columns = "contact", axis = 1, inplace = True)
```

```python
# melt auralin and novodra columns into treatment, start_dose and end_dose

id_vars = ['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change']
value_vars = ['auralin', 'novodra']
treatment_clean = pd.melt(frame = treatment_clean, id_vars = id_vars, value_vars = value_vars, var_name = 'treatment', value_name = "dose")

# select entries where dose is not -
treatment_clean = treatment_clean[treatment_clean.dose != '-']

# split dose to start and end_dose and drop dose
treatment_clean[['start_dose', 'end_dose']] = treatment_clean['dose'].str.split('-', expand = True)

# drop dose column
treatment_clean.drop('dose', axis = 1, inplace = True)
```

```python
# join adverse_clean with treatment clean
complete_treatment = pd.merge(left = treatment_clean, right = adverse_clean, on = ['given_name', 'surname'], how = 'left')

# remove given name and surname from complete_treatment dataframe and then create patient_id as a common column

# capitalize given-name and surname in complete_treatment
complete_treatment[['given_name', 'surname']] = complete_treatment[['given_name', 'surname']].apply(lambda x: x.str.capitalize())

# join patient id with complete treatment 

treatment_dff = pd.merge(left = complete_treatment, right = patient_clean[["patient_id", "surname", "given_name"]], 
         on = ["surname", "given_name"], how = 'inner')

# drop surname and given_name columns
treatment_dff = treatment_dff.drop(['given_name', 'surname'], axis = 1)

```

#### Define

* replace Californaia with CA, New York with NY, Florida  with FL, Illinois with IL and  Nebraska with NE in the state of patient_clean
* remove country column from patient_clean since all the patients are from the United States

#### Code

```python
"""replace Californaia with CA, New York with NY, Florida with FL, 
Illinois with IL and Nebraska with NE in the state of patient_clean
""" 
abbrev_state = {"California": "CA",
               "New York": "NY", 
               "Illinois": "IL", 
               "Florida": "FL", 
               "Nebraska": "NE"}

def abbreviate_state(df):
    if df['state'] in abbrev_state.keys():
        abbr = abbrev_state[df['state']]
        return abbr
    else:
        abbr = df['state']
        return abbr
    
patient_clean.loc[:, "state"] = patient_clean.apply(abbreviate_state, axis = 1)
```

```python
# remove country column from patient_clean since all the patients are from the United States

patient_clean.drop(columns = 'country', inplace = True)
```
